```python
from sklearn import preprocessing #标准化数据模块
from sklearn.model_selection import train_test_split # 将资料分割成train与test的模块
from sklearn.datasets.samples_generator import make_classification # 生成适合做classification资料的模块
from sklearn.svm import SVC  # Support Vector Machine中的Support Vector Classifier
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#建立Array
a = np.array([[10, 2.7, 3.6],
              [-100, 5, -2],
              [120, 20, 40]], dtype=np.float64)

#将normalized后的a打印出
print(preprocessing.scale(a))
```

## 例子2 - 数据标准化对机器学习成效的影响

```python
#生成具有2种属性的300笔数据
X, y = make_classification(
    n_samples=300, n_features=2,
    n_redundant=0, n_informative=2, 
    random_state=22, n_clusters_per_class=1, 
    scale=100)

#可视化数据
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
# 数据标准化前
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC()
clf.fit(X_train, y_train)
print('mean accuracy:', clf.score(X_test, y_test))
```

```python
# 数据标准化后
X = preprocessing.scale(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC()
clf.fit(X_train, y_train)
print('mean accuracy:', clf.score(X_test, y_test))
```
