# Pandas: When cell contents are lists, create a row for each element in the list

https://stackoverflow.com/a/27266225/3744499

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {'trial_num': [1, 2, 3, 1, 2, 3],
     'subject': [1, 1, 1, 2, 2, 2],
     'samples': [list(np.random.randn(3).round(2)) for i in range(6)]
    }
)
```

```python
df
```

```python
df['samples']
```

```python
res = df.set_index(['subject', 'trial_num'])['samples'].apply(pd.Series).stack().reset_index()
res
```

```python
res.columns = ['subject','trial_num','sample_num','sample']
res
```
